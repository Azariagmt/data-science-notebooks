```python
from os import system, name
from IPython.display import clear_output

playing = 'player_1'

player_symbols = {}

s = []

def clear():
    # for windows
    if name == 'nt':
        _ = system('cls')
    # for mac and linux(here, os.name is 'posix')
    else:
        _ = system('clear')
        
def set_s():
    global s
    s = [_ for _ in range(0, 11)]

def print_board(*s):

    print(f"""
         |      |
      {s[7]}  |   {s[8]}  |   {s[9]}
    _____|_____ |_____
         |      |
      {s[4]}  |   {s[5]}  |   {s[6]}
    _____|_____ |_____
         |      |
      {s[1]}  |   {s[2]}  |   {s[3]}
         |      |
    """)


def switch_turn():
    global playing
    if playing == 'player_1':
        playing = 'player_2'
    elif playing == 'player_2':
        playing = 'player_1'


def winner_check():
    if s[1] == s[2] == s[3] or s[4] == s[5] == s[6] or s[7] == s[8] == s[9]:
        return f'{playing} is the winner!!!'

def sym_choice():
    symbol_choice = input(f'Enter choice of symbol {playing}... \n ')
    if symbol_choice == 'O':
        player_symbols[playing] = 'O'
        print(f'{playing} symbol is O')
        switch_turn()
        player_symbols[playing] = 'X'
        print(f'{playing} symbol is X')
        switch_turn()
    elif symbol_choice == 'X':
        player_symbols[playing] = 'X'
        print(f'{playing} symbol is X')
        switch_turn()
        player_symbols[playing] = 'O'
        print(f'{playing} symbol is O')
        switch_turn()



set_s()

sym_choice()

while True:
    print_board(*s)
    position = int(input(f'Enter position {playing}\n'))
    if position in list(range(1,10)):
        if s[position] != 'X' and s[position] != 'O':
            s[position] = player_symbols[playing]
            if winner_check():
                print(winner_check())
                cont = input('do you wanna play another game? Yes or no?')
                if cont.upper() == 'YES':
                    set_s()
                    continue
                else:
                    break
            switch_turn()
            #normal
            clear()
            #jupyter
            clear_output()
        elif s[position]:
            print('error')
    else:
        print('correct range input...')
        continue
```
